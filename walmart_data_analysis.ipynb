# Data Cleaning

# Convert date columns

```python
print ('Converted date columns:')
print ('======================')
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
df
```

# Strip whitespaces in text columns

```python
text_cols = df.select_dtypes(include='object').columns
df[text_cols] = df[text_cols].apply(lambda x: x.str.strip())
df
```

# Descriptive Statistics

```python
print("\nðŸ“Š Descriptive Statistics:\n", df.describe())
print("\nðŸ“Š Categorical Columns Unique Values:\n")
for col in ['Category', 'State', 'City', 'Country']:
    print(f"{col}: {df[col].nunique()} unique")
```

# Summary Information

```python
print("Total Rows:", len(df))
print("\nDataset Columns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
```

# Key Metrics

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
top_customer = df['Customer Name'].value_counts().idxmax()

print("\n--- Key Metrics ---")
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Unique Orders: {total_orders}")
print(f"Top Customer: {top_customer}")
```

# Sales and Profit by Category

```python
category_summary = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
print("\nSales and Profit by Category:\n", category_summary)
```

# Monthly Sales Trend

```python
df['Order Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Order Month')['Sales'].sum()
monthly_sales
```

# Top 5 Products by Sales

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Products by Sales:\n", top_products)
```

```python
top_products.plot(kind='barh', title='Top 5 Products by Sales', figsize=(8, 4), color='green')
plt.xlabel("Sales ($)")
plt.tight_layout()
plt.savefig("top_products_by_sales.png")
plt.show()
```

# Sales and Profit by Category

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))
sns.histplot(df['Sales'], kde=True, ax=axs[0], color='blue')
axs[0].set_title("Distribution of Sales")
sns.histplot(df['Profit'], kde=True, ax=axs[1], color='green')
axs[1].set_title("Distribution of Profit")
plt.tight_layout()
plt.savefig("distribution_sales_profit.png")
plt.show()
```

## Category distribution

```python
sns.countplot(x='Category', data=df, palette='Set2')
plt.title("Product Category Distribution")
plt.xticks(rotation=45)
plt.savefig("category_distribution.png")
plt.show()
```

# Sales vs. Profit

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.savefig("sales_vs_profit.png")
plt.show()
```

# Sales and Profit by Category

```python
cat_summary = df.groupby('Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)
cat_summary.plot(kind='bar', title="Sales and Profit by Category")
plt.ylabel("Amount ($)")
plt.savefig("sales_profit_by_category.png")
plt.show()
```

# Heatmap of Correlation


```python
corr = df[['Sales', 'Quantity', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.savefig("correlation_heatmap.png")
plt.show()
```

# Profit distribution by Category and State

```python
pivot = df.pivot_table(index='State', columns='Category', values='Profit', aggfunc='sum')
pivot.fillna(0).plot(kind='bar', stacked=True, figsize=(14, 7))
plt.title("Profit by Category and State")
plt.ylabel("Profit ($)")
plt.tight_layout()
plt.savefig("profit_by_category_state.png")
plt.show()
```

# Date-Time Analysis

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%B')

monthly_sales = df.groupby(['Year', 'Month Name'])['Sales'].sum().reset_index()
monthly_sales = monthly_sales.sort_values(by=['Year', 'Month Name'])

sns.lineplot(data=monthly_sales, x='Month Name', y='Sales', hue='Year', marker='o')
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("monthly_sales_trend.png")
plt.show()
```

## Ship Delay

```python
df['Ship Delay (Days)'] = (df['Ship Date'] - df['Order Date']).dt.days
sns.histplot(df['Ship Delay (Days)'], bins=20, kde=True)
plt.title("Distribution of Shipping Delays")
plt.savefig("shipping_delay_distribution.png")
plt.show()
```
